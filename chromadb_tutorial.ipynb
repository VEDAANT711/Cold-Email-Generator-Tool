```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name = "my_collection")
```

```python
collection.add(
    documents=[
        "This document is about Amsterdam",
        "This document is about Greece"
    ],
    ids=['id1', 'id2']
)
```

```python
all_docs = collection.get()
all_docs
```

```python
documents = collection.get(ids=['id1'])
documents
```

```python
results = collection.query(
    query_texts = ['Query is about idols'],
    n_results = 2
)
results
```
